```python
import matplotlib.pyplot as plt
```

```python
with open('task1_result','rt') as fp:
    char_count = [txt for txt in fp]    
```

```python
alph_list = []

for num in range(65,91):
    alph_list.append(chr(num))
```

```python
alph_count_list = []
for eachline in char_count:
    line_split = eachline.split('\t')
    if line_split[0] in alph_list:
        count_alph = int(line_split[1].split('\n')[0])
        alph_count_list.append(count_alph)
```

```python
plt.title('Word Count in Corpus')
plt.xlabel('Alphabets')
plt.ylabel('Number of Words')
barlist = plt.bar(alph_list, alph_count_list)
for i in range(len(barlist)):
    if i%2 == 0:
        barlist[i].set_color('r')
    else:
        barlist[i].set_color('b')
plt.show()
```

```python
with open('task2_result','r') as rfp:
    word_count = [eachline for eachline in rfp]
```

```python
word_list = []
count_list = []
for eachline in word_count:
    line_split = eachline.split('\t')
    count_word = int(line_split[1].split('\n')[0])
    word_list.append(line_split[0])
    count_list.append(count_word)
unique_words = count_list[-1]
word_list = word_list[:-1]
count_list = count_list[:-1]
count_list = [(eachcount/unique_words) for eachcount in count_list]
```

```python
plt.title('Pronoun Count in Corpus')
plt.xlabel('Pronouns')
plt.ylabel('Number of Words')
barlist = plt.bar(word_list, count_list)
barlist[1].set_color('r')
barlist[2].set_color('b')
barlist[3].set_color('y')
barlist[4].set_color('b')
barlist[5].set_color('c')
barlist[6].set_color('g')
plt.show()
```
